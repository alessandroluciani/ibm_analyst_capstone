## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
QUERY = """
SELECT DatabaseDesireNextYear FROM DatabaseDesireNextYear
"""

df = pd.read_sql_query(QUERY, conn)

df_db = df.groupby(["DatabaseDesireNextYear"]).size().sort_values(ascending=False).reset_index()
df_db.columns = ["DatabaseDesireNextYear", "Total"]

df_db = df_db.loc[:4]

df_db["Total"].plot(kind="pie", labels=df_db["DatabaseDesireNextYear"], autopct='%1.1f%%')
```

```python
QUERY = """
SELECT LanguageDesireNextYear FROM LanguageDesireNextYear
"""

df = pd.read_sql_query(QUERY, conn)


df_db = df.groupby(["LanguageDesireNextYear"]).size().sort_values(ascending=False).reset_index()
df_db.columns = ["LanguageDesireNextYear", "Total"]

df_db = df_db.loc[:4]

df_db["Total"].plot(kind="pie", labels=df_db["LanguageDesireNextYear"], autopct='%1.1f%%')
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master WHERE AGE BETWEEN 30 AND 35
"""
df = pd.read_sql_query(QUERY, conn)

ages = df["Age"].unique().astype(int)
ages = list(np.sort(ages))

wwh = []
crh = []

for age in ages:    
    #print(age)
    filter = df["Age"]==age
    wwh.append(df[filter]["WorkWeekHrs"].median().astype(int))
    crh.append(df[filter]["CodeRevHrs"].median().astype(int))

width = 0.35

fig, ax = plt.subplots()
ax.bar(ages, wwh, width, label='WWH')
ax.bar(ages, crh, width, bottom=wwh, label='CRH')

ax.set_ylabel('Scores')
ax.set_title('Scores by WWH and CRH')
ax.legend()

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
QUERY = """
SELECT ConvertedComp, Age FROM master WHERE AGE BETWEEN 25 AND 30
"""
df = pd.read_sql_query(QUERY, conn)

ages = df["Age"].unique().astype(int)
ages = list(np.sort(ages))

cc = []

for age in ages:       
    filter = df["Age"]==age
    cc.append(df[filter]["ConvertedComp"].median().astype(int))    

plt.plot(ages, cc)
plt.show()

# len(ages)
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
QUERY = """
SELECT MainBranch
FROM master
"""
df = pd.read_sql_query(QUERY, conn)
df.head()

df_db = df.groupby(["MainBranch"]).size().sort_values(ascending=False)
# .reset_index()
df_db.columns = ["MainBranch", "Total"]

df_db.plot(kind="barh")
```

```python
QUERY = """
SELECT *
FROM DevType
"""
df = pd.read_sql_query(QUERY, conn)
df.head()

df_db = df.groupby(["DevType"]).size().sort_values(ascending=False)

df_db.columns = ["DevType", "Total"]

df_db.plot(kind="barh")
```
